```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({
    'figure.figsize': (4, 4),
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.spines.top': False,
    'axes.spines.bottom': False})
```

```python
import networkx as nx

#create an empty graph structure(a null graph: no nodes and no edges)
G = nx.Graph()

print('[Nodes]:', G.nodes)
print('[Edges]:', G.edges)
```

```python
# 1)One node at a time
G.add_node(1) #method of nx.Graph
print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
nx.draw_networkx(G)
```

```python
G.add_nodes_from([2, 3]) #method of nx.Graph
print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
nx.draw_networkx(G)
```

```python
G.add_nodes_from(range(4, 10))
print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
nx.draw_networkx(G)
```

```python
H = nx.path_graph(range(10, 15)) #nx.path_graph(n)
print('[Nodes]:{}\n[Edges]:{}]'.format(H.nodes,H.edges))
nx.draw_networkx(H)
```

```python
G.add_nodes_from(H)
print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
nx.draw_networkx(G)
```

```python
G1 = nx.Graph()

G1.add_node('spam')
G1.add_nodes_from('spam')

plt.figure(figsize=(7, 7))
nx.draw_networkx(G1, node_color='magenta', node_size=800)
```

```python
# 1) single edge
G.add_edge(1,2)
e = (2,3)
G.add_edge(*e)
print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
plt.figure(figsize=(10, 10))
nx.draw_networkx(G)
```

```python
# 2)Edges from any container(list, dict, set, etc.)
G.add_edges_from(H.edges)

#nodes are added automatically
G.add_edges_from([(14,15), (15,16), (15,17)])
G.add_edges_from([(3,5), (3,15), (6,15), (4,6), (6,7), (6,8), (9,13)])

print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
plt.figure(figsize=(10, 10))
nx.draw_networkx(G)
```

```python
V = G.nodes()
E = G.edges()
print('[Nodes]:{}\n[Edges]:{}]'.format(V, E))

n = G.number_of_nodes()
m = G.number_of_edges()
print('[Nodes]:{}\n[Edges]:{}]'.format(n, m))

print('Nodes#{}\'s neighbors:{}'.format(15, list(G.neighbors(15))))

print('Nodes#{}\'s neighbors:{}'.format(15, G.degree(15)))
```

```python
G.add_edge(1, 2, weight=4.7)
G.add_edges_from([(3,4),(4,5)], color='red')
G.add_edges_from([(1, 2, {'color':'blue'}), (2, 3,{'weight':8.3})])
G.add_weighted_edges_from([(6, 7, 3.0)])
G[1][2]['weight'] = 4.7
G.edges[1,2]["weight"] = 4

print(G[1])
print(G[1][2])
print(G.edges[1,2])
print(G.edges[1,2]['weight'])
print(G[1][2]['weight'])
print(G.get_edge_data(3,4))
```

```python
print(G.degree(1))
print(G.degree([1,2]))
print(dict(G.degree()).values())

for node in G.nodes:
    print(node, G.degree(node))
    
for node in G.nodes:
    if G.nodes[node]:
        print('Node#{}: [Name]:{:5s}, [Degree]:{}'.format(node, G.nodes[node]['name'], G.degree(node)))
        
for v1, v2 in G.edges:
    if 'weight' in G.edges[v1, v2]:
        print('({},{}):{}'.format(v1, v2, G.edges[v1,v2]['weight']))
```

```python
import numpy as np

degrees = sorted(dict(G.degree()).values())
hist = [degrees.count(k) for k in degrees]

# plotting
plt.figure(figsize=(5, 5))
ax = plt.subplot()
ax.plot(degrees, hist, 'ro-')
ax.legend(['Degree'])
ax.set_xlabel('Degree', fontsize=12)
ax.set_ylabel('Number of nodes', fontsize=12)
ax.set_title('Example network', fontsize =16)
ax.set_xticks(np.arange(1, 6, 1))
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
```

```python
Gw = nx.Graph()

Gw.add_edge("a", "b", weight=0.6)
Gw.add_edge("a", "c", weight=0.2)
Gw.add_edge("c", "d", weight=0.1)
Gw.add_edge("c", "e", weight=0.7)
Gw.add_edge("c", "f", weight=0.9)
Gw.add_edge("a", "d", weight=0.3)

elarge = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"]>0.5]
esmall = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"]<=0.5]

plt.figure(figsize=(6,6))
pos = nx.spring_layout(Gw) # positions of all nodes

#nodes
nx.draw_networkx_nodes(Gw, pos, node_size=700)

#edges
nx.draw_networkx_edges(Gw, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(Gw, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b",style="dashed")

#labels
nx.draw_networkx_labels(Gw, pos, font_size=20, font_family="sans-serif")
plt.axis("off")
plt.show()
```

```python
# 2)Edges from any container(list, dict, set, etc.)
G.add_edges_from(H.edges)

#nodes are added automatically
G.add_edges_from([(14,15), (15,16), (15,17), (6,16), (17, 6)])
G.add_edges_from([(3,5),(3,15), (6,15), (4,6), (6,7), (6,8), (9,13)])

print('[Nodes]:{}\n[Edges]:{}]'.format(G.nodes,G.edges))
plt.figure(figsize=(10, 10))
nx.draw_networkx(G)
```

```python
print(nx.shortest_path(G,1,7))
print(nx.shortest_path(G,16,17))
print([p for p in nx.all_shortest_paths(G, 16, 17)])
print('Diameter:', nx.diameter(G))
n, m = G.order(), G.size()
print('n: {}, m: {}'.format(n,m))
print('Average degree: {:.2f}'.format((2*m)/n))
```
